## ChatGPT의 도움으로

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler

# Step 1: Load sample data (Digits dataset)
digits = load_digits()
data = digits.data
labels = digits.target

# Inspect the data shape
print(f"Data shape (samples, features): {data.shape}")
print(f"Labels shape: {labels.shape}")
print(f"Unique labels: {np.unique(labels)}")

# Display a few sample images
fig, axes = plt.subplots(1, 5, figsize=(10, 3))
for i, ax in enumerate(axes):
    ax.imshow(digits.images[i], cmap='gray')
    ax.set_title(f"Label: {labels[i]}")
    ax.axis('off')
plt.suptitle("Example Digits from Dataset")
plt.show()

# Step 2: Preprocess data (standardization)
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Step 3: Apply t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30, max_iter=300)
data_tsne = tsne.fit_transform(data_scaled)

# Step 4: Visualize t-SNE results
plt.figure(figsize=(10, 7))
scatter = plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=labels, cmap='tab10', alpha=0.7, s=50)
plt.colorbar(scatter, label="Digit Label")
plt.title("t-SNE Visualization of the Digits Dataset")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
